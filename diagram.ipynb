```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
from matplotlib.patches import Rectangle

# Generate time series data
time = np.arange(0, 100)  # Historical time steps
true_values = np.sin(0.1 * time) + 0.5 * np.cos(0.05 * time)  # True historical values

# Future time series data
future_time = np.arange(100, 125)  # Future time steps
future_true_values = np.sin(0.1 * future_time) + 0.7 * np.cos(0.05 * future_time)

# Scenario 1: High error forecast (higher variance)
high_error_forecast = future_true_values + np.random.normal(0, 0.4, len(future_time))  # Add variance

# Scenario 2: Low error forecast (lower variance)
low_error_forecast = future_true_values + np.random.normal(0, 0.1, len(future_time))  # Add smaller variance

# Create figure and subplots
fig, axs = plt.subplots(ncols=2, figsize=(14, 5))

# Left panel: High Error Scenario
ax1 = axs[0]
ax1.plot(time, true_values, color="royalblue", linewidth=2, label="Historical True Values")
ax1.plot(future_time, future_true_values, color="green", linestyle="dashdot", linewidth=2, label="Future True Values")
ax1.plot(future_time, high_error_forecast, color="red", linestyle="dotted", linewidth=2, label="High Error Forecast")

ax1.set_xlabel("Time")
ax1.set_ylabel("True Values")
ax1.set_xlim([0, 125])
ax1.set_ylim([-2, 2])

# Highlight future forecast area
rect = Rectangle((100, -2), 25, 4, linewidth=1.5, edgecolor="black", facecolor="none")
ax1.add_patch(rect)

# Add legend
legend_elements = [
    Line2D([0], [0], color="royalblue", lw=2, label="Historical Values"),
    Line2D([0], [0], color="green", lw=2, linestyle="dashdot", label="True Values"),
    Line2D([0], [0], color="red", lw=2, linestyle="dotted", label="Forecasted Values"),
]
ax1.legend(handles=legend_elements, loc=(0.3, 0.7), fontsize=10, fancybox=True, frameon=True, facecolor="white")

ax1.text(10, 1.8, "a. LSTM Forecasting", fontsize=12, fontweight="semibold")

# Right panel: Low Error Scenario
ax3 = axs[1]
ax3.plot(time, true_values, color="royalblue", linewidth=2, label="Historical True Values")
ax3.plot(future_time, future_true_values, color="green", linestyle="dashdot", linewidth=2, label="Future True Values")
ax3.plot(future_time, low_error_forecast, color="red", linestyle="dotted", linewidth=2, label="Low Error Forecast")

ax3.set_xlabel("Time")
ax3.set_ylabel("True Values")
ax3.set_xlim([0, 125])
ax3.set_ylim([-2, 2])

# Highlight future forecast area
rect = Rectangle((100, -2), 25, 4, linewidth=1.5, edgecolor="black", facecolor="none")
ax3.add_patch(rect)
# Add legend
legend_elements = [
    Line2D([0], [0], color="royalblue", lw=2, label="Historical Values"),
    Line2D([0], [0], color="green", lw=2, linestyle="dashdot", label="True Values"),
    Line2D([0], [0], color="red", lw=2, linestyle="dotted", label="Forecasted Values"),
]
ax3.legend(handles=legend_elements, loc=(0.3, 0.7), fontsize=10, fancybox=True, frameon=True, facecolor="white")

ax3.text(10, 1.8, "b.Informer GRU Forecasting", fontsize=12, fontweight="semibold")

# Adjust subplot spacing
fig.subplots_adjust(wspace=0.4)

plt.show()
```

```python
data = pd.DataFrame({
    'Attention': ['ProbSparse', 'ProbSparse', 'Full Attention', 'Full Attention'],
    'Distillation': ['True', 'False', 'True', 'False'],
    'MAE': [0.15, 0.16, 0.16, 0.18],
    'RMSE': [0.18, 0.19, 0.19, 0.21]
})

# Create pivot tables for MAE and RMSE
mae_pivot = data.pivot(index="Attention", columns="Distillation", values="MAE")
rmse_pivot = data.pivot(index="Attention", columns="Distillation", values="RMSE")

# Construct combined annotations
annotations = mae_pivot.copy()
for row in annotations.index:
    for col in annotations.columns:
        annotations.loc[row, col] = f"MAE: {mae_pivot.loc[row, col]:.2f}\nRMSE: {rmse_pivot.loc[row, col]:.2f}"

# Create the heatmap
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(mae_pivot, annot=annotations, fmt="", cmap="coolwarm", cbar_kws={'label': 'MAE Values'})
ax.set_title("Combined Heatmap of MAE and RMSE")
plt.show()
```
